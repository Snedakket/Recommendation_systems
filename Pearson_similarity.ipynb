```python
# estimte the similarity with Pearson
import pandas as pd

OneManOut_rating = userrating[2194]

data = pd.read_csv('D:/Python_Projects/Recomendation_Systems/bars.csv')

userrating= pd.pivot_table(data=data, values='user_rating', index='user_id', columns='bars_id')
similar_to_OneManOut = userrating.corrwith(OneManOut_rating)

corr_OneManOut = pd.DataFrame(similar_to_OneManOut, columns=['PearsonR'])
corr_OneManOut.dropna(inplace=True)
corr_OneManOut.head()
```

```python
rating = pd.DataFrame(data.groupby('bars_id')['no_of_ratings'].mean())
OneManOut_corr_summary = corr_OneManOut.join(rating)

# display the most similar books, sorted with equals similarity according to the books rating
most_rated_bars=OneManOut_corr_summary.sort_values(by = ['PearsonR', 'no_of_ratings'], ascending=False).head(10)
detail = pd.merge(most_rated_bars, data, on='bars_id')
detail
```
